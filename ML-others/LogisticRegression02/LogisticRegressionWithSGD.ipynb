```
x=pd.read_csv('logistic_x.txt',sep='\ +',engine='python')
y=pd.read_csv('logistic_y.txt')
```

```
labels=y.values.astype(float).astype(int)
label=np.concatenate(labels,axis=None)
print(label)
```

```
print(type(x))
x1=x.iloc[:,0].values
x2=x.iloc[:,1].values
print(x1)
print(x2)
```

```
x['label']=label
print(x)
ax=plt.axes()
x.query('label==1').plot.scatter(x=0,y=1,ax=ax,color='red')
x.query('label==-1').plot.scatter(x=0,y=1,ax=ax,color='green')
print(x.shape)
```

```
for index in range(len(label)):
  if label[index]==-1:
    label[index]=0
x1=np.array(x1).reshape(x.shape[0],1)
x2=np.array(x2).reshape(x.shape[0],1)
xs=np.hstack([np.ones([x.shape[0],1]),x1,x2])
print(xs)
```

```
beta=np.ones(xs.shape[1])
print(label[1])
print(xs[1])
print(label)
```

```
thetas=[]
lr=0.001
iteration=0
totalloss=1e6
theta=np.ones(xs.shape[1])
while totalloss>1e-3:
  old_theta=theta.copy()
  for i in range(xs.shape[0]):
    #for j in range(xs.shape[1]):
    hx=1/(1+np.exp(-theta.dot(xs[i,:].T)))
    for j in range(xs.shape[1]):
      delta=lr*(label[i]-hx)*xs[i][j]
      theta[j]+=delta
  totalloss=np.sum(np.abs(theta-old_theta))
  if iteration%100==0:
    thetas.append(theta.copy())
  iteration+=1
print(iteration)
print(thetas)
```

```
ax=plt.axes()
x.query('label==1').plot.scatter(x=0,y=1,ax=ax,color='red')
x.query('label==-1').plot.scatter(x=0,y=1,ax=ax,color='green')
xtemp=np.array([np.min(xs[:,1]),np.max(xs[:,1])])
for k,theta in enumerate(thetas):
  ytemp=(theta[0]+theta[1]*xtemp)/(-theta[2])
  plt.plot(xtemp,ytemp,lw=0.5,label='iteration{0}'.format(k*100))
plt.legend(bbox_to_anchor=(1.04,1))
```
